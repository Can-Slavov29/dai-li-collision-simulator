```python
import pygame
from sys import exit
import numpy as np
```

```python
pygame.init()

#creating surfaces
screen = pygame.display.set_mode((800,600))
pygame.display.set_caption("Avatar:Battle Field")
rock1 = pygame.image.load("rock.png")
rock2 = pygame.image.load("rock.png")
background = pygame.image.load("background.png")


clock = pygame.time.Clock()

#creating rock surfaces
radius = 10
rock1 = pygame.Surface((2*radius,2*radius),pygame.SRCALPHA)
pygame.draw.circle(rock1,"brown",(radius,radius),radius)
rock2 = pygame.Surface((2*radius,2*radius),pygame.SRCALPHA)
pygame.draw.circle(rock2,"brown",(radius,radius),radius)

#randomising x and y positions
x1 = np.random.choice(np.arange(100,700,1)) + radius
x2 = np.random.choice(np.arange(100,700,1)) + radius
y1 = np.random.choice(np.arange(100,500,1)) - radius
y2 = np.random.choice(np.arange(100,500,1)) - radius

#randomising x and y velocities
vx1 = np.random.choice(np.arange(-400,400,1)) 
vx2 = np.random.choice(np.arange(-400,400,1)) 
vy1 = np.random.choice(np.arange(-400,400,1)) 
vy2 = np.random.choice(np.arange(-400,400,1)) 

                      

#initialisg velocity and position
pos_rock1 = np.array([x1,y1])
pos_rock2 = np.array([x2,y2])

v_rock1 = np.array([vx1,vx2])
v_rock2 = np.array([vx2,vy2])



test_surface = pygame.Surface((100,200))


while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            exit()
    screen.blit(background,np.array([0,0]))
    screen.blit(rock1,(pos_rock1[0]-radius,pos_rock1[1]-radius))
    screen.blit(rock2,(pos_rock2[0]-radius,pos_rock2[1]-radius))
    

    
    pos_rock1 = pos_rock1 + v_rock1*clock.tick(60)/1000
    pos_rock2 = pos_rock2 + v_rock2*clock.tick(60)/1000
    
    

    rel_pos= pos_rock1 - pos_rock2
    distance = np.linalg.norm(rel_pos)

    if not 100+radius<=pos_rock1[0]<=700-radius:
        v_rock1[0] = v_rock1[0]*-1

    if not 100+radius<=pos_rock2[0]<=700-radius:
        v_rock2[0] = v_rock2[0]*-1

    if not 100+radius<=pos_rock1[1]<=500-radius:
        v_rock1[1] = v_rock1[1]* -1

    if not 100+radius<=pos_rock2[1]<=500-radius:
        v_rock2[1] = v_rock2[1]* -1
    
    
        

        
        

    
    pygame.display.update()
    clock.tick(60)
```
