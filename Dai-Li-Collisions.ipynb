```python
def create_rocks(nmax,radius):
        rocks = []
        for i in range(0,nmax):
            #creating rock surface
            rock = pygame.Surface((2*radius,2*radius),pygame.SRCALPHA)
            pygame.draw.circle(rock,"brown",(radius,radius),radius)
            rocks.append(rock)
        return rocks

def posvel_rocks(nmax,radius):
    pos_x = []
    pos_y = []
    vel_x = []
    vel_y = []
    
    for n in range(0,nmax):  
        xp = np.random.choice(np.arange(100+radius,700-radius,1)) 
        yp = np.random.choice(np.arange(100+radius,500-radius,1)) 

        vx = np.random.choice(np.arange(-400,400,1))
        vy = np.random.choice(np.arange(-400,400,1))

        
        
        pos_x.append(xp)
        pos_y.append(yp)

        vel_x.append(vx)
        vel_y.append(vy)
    
    return np.array([pos_x,pos_y]),np.array([vel_x,vel_y])
    
    
pos, vel = posvel_rocks(2,10)

pos

print(pos)
print(vel)

pos+vel*0.5


```
